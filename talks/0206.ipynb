# 206. Reverse Linked List

## Question
* https://leetcode.com/problems/reverse-linked-list/

## Solutions
* Iteration
    * copied or in place
    * Time: O(N)
    * Space: O(N)

* Recursion
    ```
    # CnPC = PCCn
	cur.next, prev, cur = prev, cur, cur.next
    ```
    * In place
    * Time: O(N)
    * Space: O(1)

````python
from typing import List


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution0:
	'''
    Iteration in place
	'''

	def reverseList(self, head: ListNode) -> ListNode:
		cur, prev = head, None
		while cur:
            # CnPC = PCCn
		    cur.next, prev, cur = prev, cur, cur.next
		return prev
	

class Solution:
    '''
    iteration: with copy
    '''

    def reverseList(self, head: ListNode) -> ListNode:
        lead = cur = None
        while head:
            lead = ListNode(head.val)
            lead.next = cur
            cur = lead
            head = head.next
        return lead


class Solution2:
    '''
    recursion
    '''

    def reverseList(self, head: ListNode) -> ListNode:
        if not head or not head.next:
            return head
        cur = self.reverseList(head.next)
        head.next.next = head
        head.next = None
        return cur


def print_list(node: ListNode):
    if not node:
        print()
        return
    print(node.val, end=',')
    print_list(node.next)


def ll_to_list(head: ListNode) -> List[int]:
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result


ll = ListNode(1, next=ListNode(2, next=ListNode(
    3, next=ListNode(4, next=ListNode(5)))))

s = Solution0()
assert ll_to_list(s.reverseList(ll)) == [5, 4, 3, 2, 1]

ll = ListNode(1, next=ListNode(2, next=ListNode(
    3, next=ListNode(4, next=ListNode(5)))))
s = Solution()
assert ll_to_list(s.reverseList(ll)) == [5, 4, 3, 2, 1]

ll = ListNode(1, next=ListNode(2, next=ListNode(
    3, next=ListNode(4, next=ListNode(5)))))

s2 = Solution2()
assert ll_to_list(s2.reverseList(ll)) == [5, 4, 3, 2, 1]

print("pass")
````
